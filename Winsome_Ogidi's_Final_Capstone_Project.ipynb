## **Data Analysis**

---

**Performing analysis to discover insights and trends**

Calculating and adding the following to the DataFrame:
*   Discount Amount
*   Net Booking Fee
*   Tax Paid
*   Booking Fee After Tax
*   Profit
*   Age Group




This is all done to know the profit made by the Hotel











```python
# Discount Amount

Hotel_Bookings_Data["Discount Amount"] = (Hotel_Bookings_Data["Booking Price[SGD]"] * Hotel_Bookings_Data["Discount"]) / 100

#let's see it in the DataFrame
Hotel_Bookings_Data
```

```python
# Net Booking Fee considering the discount which would be deducted from the initial booking fee

Hotel_Bookings_Data["Net Booking Fee"] = Hotel_Bookings_Data["Booking Price[SGD]"] - Hotel_Bookings_Data["Discount Amount"]
```

```python
# Tax Paid
#GST is the indirect tax imposed

Hotel_Bookings_Data["Tax Paid"] = Hotel_Bookings_Data["GST"] * Hotel_Bookings_Data["Booking Price[SGD]"]
```

```python
# Booking Fee After Tax
#this is to calculate the total booking charges in the hotel with tax inclusive

Hotel_Bookings_Data["Booking Fee After Tax"] = Hotel_Bookings_Data["Net Booking Fee"] - Hotel_Bookings_Data["Tax Paid"]
```

```python
# Profit
#determining the actual profit from bookings by the hotel

Hotel_Bookings_Data["Profit"] = (Hotel_Bookings_Data["Booking Fee After Tax"] * Hotel_Bookings_Data["Profit Margin"]) / 100
```

```python
# Age Group
#placing the ages in age group
Hotel_Bookings_Data["Age Group"] = pd.cut(Hotel_Bookings_Data["Age"],[19,31,41,51, np.inf], labels = ["Below 31", "31-40","41-50","Above 50"], right = False)
```

```python
# Now let's see all the new calculated columns added
Hotel_Bookings_Data.head(10)
```
